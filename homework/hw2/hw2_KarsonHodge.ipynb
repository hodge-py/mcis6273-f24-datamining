```python
def generate_gutenberg_dataset(guten_ids, outfile):
    import requests
    import time

    data = ""

    for idf in guten_ids:
        url = f"https://www.gutenberg.org/ebooks/{idf}.txt.utf-8"
        print(url)
        with requests.get(url) as r:
            data = data + r.content.decode("utf-8")
            print(f"[info]: data obtained from {url}")
            time.sleep(10)

    with open(outfile, "w", encoding="utf-8") as fo:
        fo.write(data)
```

```python
## YOUR CODE GOES HERE

herodotus = [2707, 2131, 2456]
lewis_carroll = [11, 12, 13, 23717]
f_s_fitzgerald = [805, 6695, 64317]
t_smollett = [6761, 2160, 4085]
```

```python
## THIS CODE SHOULD NOT BE CHANGED!

generate_gutenberg_dataset(herodotus, "herodotus.txt")
generate_gutenberg_dataset(lewis_carroll, "lewis_carroll.txt")
generate_gutenberg_dataset(f_s_fitzgerald, "f_s_fitzgerald.txt")
generate_gutenberg_dataset(t_smollett, "t_smollett.txt")
```

```python
test_train_documents = [
    'herodotus.txt', 't_smollett.txt', 'f_s_fitzgerald.txt', 'lewis_carroll.txt', "herodotus_test.txt", "t_smollett_test.txt",  "f_s_fitzgerald_test.txt", "lewis_carroll_test.txt"
    ]
    
test_classes = [1, 2, 3, 4] ## DO NOT CHANGE
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(input="filename", stop_words="english")
```

```python
## COMPLETE THIS

X = vectorizer.fit_transform(test_train_documents) # you will need to set the vectorized to be ALL documents
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
y = test_classes

clf.fit(X[:4], y)

### DO THIS: call the fit() method on the first 4 documents of X and the classes in y
```

```python
## DO NOT CHANGE!
clf.predict(X[4:])
```

```python
from sklearn.metrics import classification_report
def warn(*args, **kwargs): # Hide warmings for classification report
    pass
import warnings
warnings.warn = warn
print(classification_report(test_classes, clf.predict(X[4:])))
```

### The multinomialNB classifier is quite accurate because it was able to classify three of the four classes correctly. With Class 3 accidentally being attributed to class 4 and therefore having a precision rate of 50% but its recall (correct true positives in the sample) is 100%.
