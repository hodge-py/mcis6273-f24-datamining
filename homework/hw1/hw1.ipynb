```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("en.openfoodfacts.products.cookies.csv", on_bad_lines='skip',header=0,delimiter="\t")
df.head()
```

```python
df2 = df[['countries_en', 'completeness', 'serving_size', 'energy-kcal_100g', 'fat_100g', 'saturated-fat_100g', 'trans-fat_100g', 'cholesterol_100g', 'carbohydrates_100g', 'sugars_100g', 'fiber_100g', 'proteins_100g', 'sodium_100g', 'vitamin-a_100g', 'vitamin-d_100g', 'vitamin-e_100g', 'vitamin-k_100g', 'vitamin-c_100g', 'vitamin-b1_100g', 'vitamin-b2_100g', 'vitamin-pp_100g', 'vitamin-b6_100g', 'vitamin-b9_100g', 'folates_100g', 'vitamin-b12_100g', 'potassium_100g', 'calcium_100g', 'iron_100g', 'magnesium_100g', 'zinc_100g', 'copper_100g', 'manganese_100g', 'iodine_100g', 'caffeine_100g']]
df2.head()
```

```python
countries = ['United States', 'France', 'Spain', 'United Kingdom', 'Canada', 'Italy', 'Australia', 'Switzerland', 'Brazil', 'India']
df3 = df2[df2['countries_en'].isin(countries)]
df3.head()
```

```python
df4 = df3[df3['completeness'] > .60]
df4.head()
```

```python
df5 = df4[df4['energy-kcal_100g'] > 0.0]
df5.head()
```

```python
df5.to_csv("cookies.data-filtered.csv")
```

```python
country_val = df5['countries_en'].value_counts()
country_val.plot.bar(x=country_val)
```

```python
bins = pd.cut(df5['completeness'],5)
bins.value_counts().plot.bar()
```

```python
df5['serving_size'] = df5['serving_size'].str.replace('g','')
df5['serving_size'] = df5['serving_size'].str.replace(',','.')
df5['serving_size'] = pd.to_numeric(df5['serving_size'],errors='coerce')
df5 = df5[df5['serving_size'].notna()]
df5.head()
```
